#### Problem 1: Mobile Recharge Validation System
Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps


Payment gateways


Hint
Use function for validation


Use while loop for retry

```python

def mobile_recharge_req():
    valid_plan=[199, 299, 399, 599]
    while True:
        try:
            recharge=int(input('Enter recharge amount'))
            if recharge<50:
                print('Enter Valid amount')
                continue
            print('Charge Plan:\nRs.199\nRs.299\nRs.399\nRs.599\n')
            n=int(input('\nSelect the plan:'))
            if n not in valid_plan:
                print('Invalid plan,please enter valid one')
                continue
            
            print(f'Recharge succesful of Rs.{n} plan')
            break 
        except ValueError:
            print('Invalid,try again')

res=mobile_recharge_req()
print(res)
```

#### Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use
Warehouse systems


Retail management


Hint
Use function


Use for loop on dictionary

```python
def inventory_alert(storage):
    for product,stock in storage.items():
        if stock<15:
            print(f'Reorder alert for {product}')
        else:
            print(f'Stock ok for {product}')

Storage={'Laptop':3,'Mouse':20,'Keyboard':10,'Charger':16}
res=inventory_alert(Storage)
```

#### Problem 3: Student Result Processing System
Problem Statement
Create a function that:
Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint
Use loop to calculate total


Return result from function


```python
def process_result(marks):
    total = 0
    
    for i in marks:
        total += i
    
    avg = total / len(marks)
    print(f"Average Marks: {round(avg, 2)}")
    
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"


marks = [67, 69, 47, 93, 91, 87, 77]
result = process_result(marks)
print("Result:", result)
```

#### Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


Real-Time Use
Ride-hailing apps


Transport systems


Hint
Function for fare calculation


While loop for retry


```python
def cab_fare_estimator():
    while True:
        try:
            base_fare=50
            distance=int(input('Enter distance travelled'))
            if distance < 0:
                print("Distance cannot be negative. Try again.\n")
                continue

            total_fare=base_fare+distance*12
            peak_hour=input('Peak Hour?(yes or no)').strip().lower()
            
            if peak_hour == 'yes':
                total_fare+=0.25*total_fare
                
            print(f'Total fare is Rs.{round(total_fare,2)}')
            
            retry=input('Do you want to check again(yes or no)').strip().lower()
            if retry == 'yes':
                continue
            elif retry == 'no':
                return "Thank you for using Cab Fare Estimator"
            else:
                print('Invalid input. Please type yes or no.\n')
                
        except ValueError:
            print("Please enter valid numeric distance.\n")

res=cab_fare_estimator()
print(res)
```

#### Problem 5: Employee Attendance Eligibility Checker

Create a function that:
Accepts attendance list ("P", "A")


Calculates attendance percentage


Returns Eligible if ≥ 75% else Not Eligible


Real-Time Use
HR payroll systems


Internship tracking


 Hint
Use loop to count present days


Use return statement


```python
def check_attendance(att):
    p = 0
    
    for i in att:
        if i == "P":
            p += 1

    perc = (p / len(att)) * 100
    print("Attendance Percentage:", perc)

    if perc >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


emp1 = ["P", "P", "A", "P", "P"]
emp2= ["A","A","A","P","P"]
emps=[emp1,emp2]
for emp in emps:
    result = check_attendance(emp)
    print("Status:", result)
```

#### Problem 6: Password Strength Checker
Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8


Must contain at least one digit


Must contain at least one special character (@#$)


Real-Time Use
Login systems


Security validation


Hint
Use loop over string


Use conditions inside function


```python
def password_checker():
    pwd=input('Enter the password')
    if len(pwd)<8:
        return 'Invalid password,Minimum length=8'
        
    for i in pwd:
        if i.isdigit() is True:
            break
    else:
        return 'Invalid password,no digit present'
        
    for i in pwd:
        if i in '@#$':
            break
    else:
        return 'Invalid passwrod,must contain one special character'

    return f'{pwd}\nPassword is valid and strong'

res=password_checker()
print(res)
```
