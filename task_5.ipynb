#### Problem 1: Social Media – Post Engagement Analyzer
Problem Statement
Create a function to analyze engagement on a social media post.
Rules
Calculate total likes
If Total Likes ≥ 1000 → Viral Post
Otherwise → Normal Engagement
Real-Time Use
Social media analytics platforms
Marketing dashboards


```python
def post_engage(likes):
    total_likes=0
    
    #Counting likes
    for like in likes:
        total_likes+=like
        
    #Checking status 
    if total_likes >= 1000:
        print(f"Total likes: {total_likes}\nPost Status: Viral Post")
    else:
        print(f"Total likes: {total_likes}\nPost Status: Normal Engagement")
        
likes = [220, 320, 450, 11]   # Sample like counts
post_engage(likes)
```

#### Problem 3: Agriculture – Rainfall Adequacy Checker
Problem Statement
Create a function that analyzes seasonal rainfall data.
Rules
Calculate average rainfall
Average ≥ required level → Adequate Rainfall
Otherwise → Inadequate Rainfall
Real-Time Use
Climate monitoring systems
Irrigation planning tools


```python
def rainfall_adeq(rainfall,required):
    total_rainfall=0
    
    #Calculating average rainfall
    for rain in rainfall:
        total_rainfall+=rain
    avg=total_rainfall/len(rainfall)
    print('Average Rainfall:',int(avg))

    #Checking rainfall status
    if avg>=required:
        print('Rainfall Status: Adequate Rainfall')
    else:
        print('Rainfall Status: Inadequate Rainfall')
        
season_rainfall=[70, 75, 68, 75]
required_level=67 
rainfall_adeq(season_rainfall,required_level)
```

#### Problem 4: Social Media – Duplicate Account Detection
Problem Statement
Create a function to detect duplicate usernames in a system.
Rules
If duplicates exist → Flag accounts
Else → Accounts are unique
Real-Time Use
Platform moderation
Fraud detection systems


```python
def detect_duplicate(usernames):
    new_user=set(usernames)
    
    #Comparing the new username collection of set type to old one to check for duplicates
    if len(usernames)==len(new_user):
        print('Duplicate Accounts Found: No')
    else:        
        print('Duplicate Accounts Found: Yes')
               
users=['Nikhil','Knight','Rook','Nikhil']
detect_duplicate(users)
```

#### Problem 6: Agriculture – Premium Crop Price Filter
Problem Statement
Create a function that filters crop prices above a premium threshold.
Rules
Price > ₹2000 → Premium Crop
Else → Regular Crop
Real-Time Use
Market analytics
Export pricing systems


```python
def premium_crop_filter(crops):
    
    #To store premium crop price 
    premium_crops=[]
    for price in crops:
        if price>2000:
            premium_crops.append(price)
            
    print(f'Premium Crops:{premium_crops}')
    
Crops=[1000,2000,2500,1100,3200]
premium_crop_filter(Crops)
```

Create a function to evaluate application health based on error count.
Rules
Errors = 0 → Healthy
Errors ≤ 5 → Minor Issues
Errors > 5 → Critical Issues
Real-Time Use
System monitoring tools
IT operations dashboards


```python
def health_checker(errors):
    #Using if elif else conditions to assigne status
    if errors==0:
        Status='Healthy'
    elif errors<=5:
        Status='Minor Issues'
    elif errors>=5:
        Status='Critical Issues'
    else:
        print('Invalid input')
        
    print(f'Error Count: {errors}')
    print(f'System Status: {Status}')

Errors=7
health_checker(Errors)
```

#### Problem 9: E-Learning – Student Attendance Eligibility System
Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.
Rules
Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible
Real-Time Use
Learning management systems
Internship tracking platforms


```python
def attend_eligible(atd):
    p=0
    for i in atd:
        i=i[0].lower() #to get only the first character in lowercase for efficient comparison and calculation
        if i=='p':
            p+=1
        
    perc=(p/len(atd))*100 
    
    #assigning status
    if perc >= 75:
        status='Eligible'
    else:
        status='Not Eligible'
    
    print(f'Attendance Percentage: {round(perc,2)}') #rounding up for cleaner format
    print(f'Exam Eligibility: {status}') 
    
Attendance=['P','Absent','present','P','Presnt']
attend_eligible(Attendance)
    
```

#### Problem 10: Smart Electricity Bill Analyzer
Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.
Billing Rules
First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit
Usage Classification
Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage
Real-Time Use
Smart meter systems
Electricity board billing software
Energy consumption analytics


```python
def usage_level(units):
    print('Units:',units)
    
    # calculating electricity bill(slab-wise)
    if 0<units<=100:
        bill=units*3
    elif 101<=units<=200:
        bill=300+(units-100)*5
    elif units>200:
        bill=300+500+(units-200)*7
    else:
        print('Invalid Input')
        return #function terminates for invalid input.
    
    print(f'Total bill= Rs.{bill}')
    
    #usage category based on bill amount.
    if bill<500:
        print('Low Usage')
    elif 500<=bill<=1500:
        print('Moderate Usage')
    elif bill>1500:
        print('High Usage')
    else:
        return   
    
Units=int(input("Enter no of units(must be above 0)"))
usage_level(Units)
```
