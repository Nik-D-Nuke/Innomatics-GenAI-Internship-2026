#### Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {"Ravi": 92,"Anita": 88,"Kiran": 92,"Suresh": 85}

#identifies the highest performance score.
max_sal=max(employees.values())

#storing employees sharing the highest salary to handle ties
max_emp=[]
for i,j in employees.items():
    if max_sal==j:
        max_emp.append(i)

#displaying the employees eligible for top performance bonus
print(f"Top Performers Eligible for Bonus: {', '.join(max_emp)} (Score:{max_sal})")
```

#### Problem Statement 2: Search Query Keyword Analysis

```python
import string 

query="Buy mobile phone ,buy phone online"
 
#lowercase.
query=query.lower()

#Ignore common punctuation.
for p in string.punctuation:
    query=query.replace(p,"")

#count the frequency of each keyword.
query=query.split(" ")
count={}
for i in query:
    if i not in count:
        count[i]=1
    else:
        count[i]+=1

#Display only keywords searched more than once.
cnt={}
for i,j in count.items():
    if j==2:
        cnt[i]=j
cnt
```

#### Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_reading=[]

# Store even readings as (hour_index, reading_value) pairs
for i,j in enumerate(sensor_readings):
    if j%2==0:
        valid_reading.append((i,j))
    
print('Valid Sensor Readings (Hour, Value):\n',valid_reading)
```

```python
#using list comprehension
valid_readings=[(i,j) for i,j in enumerate(sensor_readings) if j%2==0]
print('Valid Sensor Readings (Hour, Value):')
print(valid_reading)
```

#### Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

#for count and percentage
n=len(emails)
g=y=0

#count of each domain
for e in emails:
    #for breaking and selecting the domain part from the string to compare
    domain = e.split("@")[1]
    if domain=='gmail.com':
        g+=1
    elif domain=='yahoo.com':
        y+=1
 
#percentage calculation
gmail=int((g/n)*100)
yahoo=int((y/n)*100)
print(f'gmail.com: {gmail}%\nyahoo.com: {yahoo}%')
```

#### Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

#average
avg=sum(sales)/len(sales)
threshold = avg * 1.3

#days where sales are more than 30% above average.
for s in sales:
    if s-0.3*s>avg:
        print(s)
        
for i,s in enumerate(sales):
    if s > threshold:
        print(f"Day {i+1}: {s}")
```

#### Problem Statement 6: Duplicate User ID Detection

```python
from collections import Counter

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

counts = Counter(user_ids)

for user, count in counts.items():
    if count > 1:
        print(f"{user} → {count} times")
```
