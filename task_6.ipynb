#### Problem Statement 1:Smart Parking Lot Management System
Design a function to manage a smart parking lot.
The system should:
Accept vehicle entry and exit logs


Calculate total parked vehicles


Identify peak parking usage


Alert if parking exceeds capacity


 Real-Time Use
Mall parking systems


Smart city infrastructure


```python
def smart_parking_system(vehicle_log,capacity):
    current_parked=0
    peak_usage=0
    
    # counting based on IN and OUT logs    
    for log in vehicle_log:
        if log=='IN':
            current_parked+=1
        elif log=='OUT':
            if current_parked>1:
                current_parked-=1
                
    #assigning peak usage 
    if current_parked>peak_usage:
        peak_usage=current_parked
        
    print("Currently Parked Vehicles:", current_parked)
        
    # checking parking status based on capacity
    if current_parked < capacity:
        print("Parking Status: Available")        
    elif current_parked == capacity:
        print("Parking Status: Full")
    else:
        print("ALERT: Parking capacity exceeded!")
        
    print("\nPeak Parking Usage:", peak_usage)
        
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(vehicle_logs,capacity)
```

#### Problem Statement 2:Online Food Delivery Time Estimator
Create a function that estimates delivery time based on:
Distance (km)


Weather condition


Traffic level


Apply delays dynamically and display final ETA.
Real-Time Use
Food delivery apps


Logistics platforms


```python
def estimate_delivery_time(distance, traffic, weather):

    # Adjust speed based on traffic
    if traffic == "Low":
        speed = 40
    elif traffic == "Medium":
        speed = 25
    elif traffic == "High":
        speed = 12
    else:
        speed = 30   
        # default speed

    # Calculate base time
    time = (distance / speed) * 60

    # Apply weather delay
    if weather == "Rainy":
        time += 15
    elif weather == "Stormy":
        time += 25

    return round(time)


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

eta = estimate_delivery_time(distance, traffic, weather)
print("Estimated Delivery Time:", eta, "minutes")

    
```

####  Problem Statement 5:Smart Classroom Resource Usage Monitor
Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.
Real-Time Use
Smart classrooms


Energy optimization systems


```python
def resource_usage_monitor(resource_usage):
    overused=[]
    alert='No'
    
    #overused resources (more than 8 hours)
    for resource,hours in resource_usage.items():
        if hours>8:
            overused.append(resource)
            alert='Yes'
            
    if overused:
        print('Overused Resources:',", ".join(overused))
        print(f'Energy alert: {alert}')
    
    # calculate total usage
    total_hrs=sum(resource_usage.values())
    print('\nTotal usage hours of all resources:',total_hrs)
    
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4}
resource_usage_monitor(resource_usage)
```

####  Problem Statement 6: Online Event Registration Capacity Controller
Create a function that manages event registrations by:
Tracking registrations


Preventing overbooking


Triggering waitlist mode


 Real-Time Use
Webinar platforms


Conference registration systems


```python
def registration_controller(capacity, registrations):
    
    confirmed = 0
    waitlisted = 0
    
    # track confirmed and waitlisted users
    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    #closing registration if capacity is reached
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


capacity = 100
registrations = 105

registration_controller(capacity, registrations)
```
